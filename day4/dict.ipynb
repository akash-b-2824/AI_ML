```python
contacts = {
    "akash": "555-1234",
    "Boby": "555-2345",
    "monika": "555-3456"
}

contacts["akash"] = "555-4567"   
contacts["monika"] = "555-9999"     

existing_lookup = contacts.get("akash", "Contact not found")
missing_lookup = contacts.get("ananya", "Contact not found")

for name, phone in contacts.items():
    print(f"Contact: {name} | Phone: {phone}")

print("\nLookup existing:", existing_lookup)
print("Lookup missing:", missing_lookup)
```

```python
raw_logs = ["ID01", "ID02", "ID01", "ID05", "ID02", "ID08", "ID01"]
unique_users = set(raw_logs)
print("Unique users:", unique_users)
print('"ID05" is unique users ?', "ID05" in unique_users)
print("Original count:", len(raw_logs))
print("Unique count:", len(unique_users))
```

```python
akku= {"Python", "Cooking", "Hiking", "Movies"}
pakku = {"Hiking", "Gaming", "Photography", "Python"}
shared = akku & pakku
alla = akku | pakku
unique = akku - pakku
print("Shared interests:", shared)
print("All interests:", alla)
print("Unique to friend_a:", unique)
```
